---

## 📘 O que é o Databricks?

O **Databricks** é uma plataforma unificada de **dados e inteligência artificial (IA)** desenvolvida para simplificar a engenharia de dados, ciência de dados, análise e machine learning em **ambientes de big data**.

A plataforma foi criada pelos criadores do **Apache Spark**, e é amplamente utilizada por empresas que precisam trabalhar com grandes volumes de dados de forma **colaborativa, escalável e segura**.

---

## 🧱 Componentes Principais

### 1. Apache Spark
- Motor de processamento distribuído para dados em larga escala.
- Suporte a múltiplas linguagens: **SQL, Python, Scala, R**.

### 2. Notebooks Colaborativos
- Ambiente interativo para escrever e executar código.
- Suporte a Markdown, visualizações de dados, e comentários em tempo real.

### 3. Delta Lake
- Camada de armazenamento transacional ACID sobre data lakes.
- Permite operações de leitura e escrita seguras, versionamento e time travel.

### 4. MLflow
- Ferramenta para **gerenciar o ciclo de vida de modelos de machine learning**.
- Inclui rastreamento de experimentos, versionamento de modelos e deploy.

### 5. Jobs e Pipelines
- Criação de pipelines de dados (ETL/ELT).
- Agendamento de jobs para processamento automatizado.

---

## ☁️ Integração com Nuvem

O Databricks é um **serviço gerenciado em nuvem**, disponível nas três principais plataformas:

- **Azure Databricks**  
  Integração com Azure Data Lake, Synapse, Key Vault etc.

- **AWS Databricks**  
  Integração com S3, Redshift, IAM e outros serviços da AWS.

- **Google Cloud Databricks**  
  Integração com BigQuery, Cloud Storage, IAM do GCP.

## Criando nosso primeiro Cluster Spark

## No Databricks, um cluster é um conjunto de recursos computacionais (máquinas virtuais) que executa os códigos que você escreve nos notebooks — como transformações de dados, consultas SQL, treinos de modelos de machine learning, entre outros.

🧠 Em termos simples:
Um cluster no Databricks é como o "cérebro" por trás das suas operações com dados. É onde o Spark roda e onde o processamento realmente acontece.

🔧 Componentes do Cluster
Driver Node: Gerencia o cluster, coordena a execução das tarefas e mantém o notebook conectado.

Worker Nodes: Executam a maior parte do trabalho — transformações, cálculos e armazenamentos temporários.

💡 O Databricks usa o Apache Spark como motor de execução distribuído, por isso os clusters são essenciais para processar grandes volumes de dados de forma eficiente.

## ⚙️ Tipos de Clusters
Tipo de Cluster	Descrição
### Interactive Cluster	Usado para notebooks, testes, análises ad hoc. Você interage com ele manualmente.
###  Job Cluster	Criado automaticamente para executar tarefas agendadas (jobs), e é encerrado ao final.
###  All-purpose Cluster	Usado por vários usuários e notebooks ao mesmo tempo. Ideal para colaboração.

## O que são Workspaces?
Os Workspaces são espaços de trabalho virtuais que permitem que você crie e execute notebooks, scripts e outros recursos relacionados ao seu projeto. Eles são organizados em uma hierarquia de pastas e podem ser compartilhados com outros usuários.

## Como criar um Workspace?

Ao importar um arquivo para o Databricks, você pode criar um novo Workspace ou escolher um existente. Para criar um novo Workspace, siga estas etapas:

1. Clique no botão "Novo Workspace" no menu principal.
2. Selecione a opção "Criar um novo Workspace".
3. Digite um nome para o novo Workspace e clique em "Criar".
