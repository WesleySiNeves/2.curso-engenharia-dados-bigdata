# Web Scraping com Beautiful Soup e Requests
### Web scraping é uma técnica para extrair dados de websites de forma automatizada. Vamos explorar como usar as bibliotecas Beautiful Soup e Requests para realizar web scraping em Python.

## Introdução ao Web Scraping
Web scraping é o processo de coletar informações de páginas web de forma automatizada. É útil para:

- Coleta de dados para análise
- Monitoramento de preços
- Pesquisa de mercado
- Agregação de conteúdo

## Bibliotecas Principais

### Requests
A biblioteca Requests é usada para fazer requisições HTTP em Python. Ela permite que você envie solicitações HTTP/1.1 de forma extremamente simples.

### Beautiful Soup
Beautiful Soup é uma biblioteca Python para extrair dados de arquivos HTML e XML. Ela cria uma árvore de análise a partir do documento HTML, o que facilita a navegação, busca e modificação da árvore.

```python
# Importando a biblioteca Requests
import requests
import json  # Adicionando biblioteca json para melhor formatação das respostas
```

# 1. Fazendo uma requisição GET simples
# O método GET é usado para solicitar dados de um recurso específico

```python
print("1. Requisição GET Básica")
# https://www.cnnbrasil.com.br/
# https://httpbin.org/get
response = requests.get('https://httpbin.org/get')
```

```python
# Verificando o status da resposta
print(f"Código de status: {response.status_code}")  # 200 significa sucesso
```

```python
response
```

```python
# Exibindo o conteúdo da resposta
print("Conteúdo da resposta:")
# Formatando a resposta JSON para melhor visualização
print(json.dumps(response.json(), indent=4))  # Melhor formatação para JSON
print("\n")
```

# 2. Enviando parâmetros em uma requisição GET
# Parâmetros são adicionados à URL após o símbolo '?'

```python
print("2. Requisição GET com Parâmetros")
payload = {'chave1': 'valor1', 'chave2': 'valor2'}
response = requests.get('https://httpbin.org/get', params=payload)
response.status_code
```

```python
# Exibindo a URL construída
print(f"URL construída: {response.url}")
print("Resposta:")
print(json.dumps(response.json(), indent=4))  # Melhor formatação para JSON
print("\n")
```

# 3. Fazendo uma requisição POST
# O método POST é usado para enviar dados ao servidor

```python
print("3. Requisição POST")
payload = {'nome': 'João', 'idade': 30}
response = requests.post('https://httpbin.org/post', data=payload)

print(f"Código de status: {response.status_code}")
print("Resposta:")
print(json.dumps(response.json(), indent=4))  # Melhor formatação para JSON
print("\n")
```

# 4. Enviando dados JSON em uma requisição POST

```python
print("4. Requisição POST com JSON")
payload = {'nome': 'Maria', 'idade': 25, 'habilidades': ['Python', 'Data Science']}
response = requests.post('https://httpbin.org/post', json=payload)

print("Resposta:")
print(json.dumps(response.json(), indent=4))  # Melhor formatação para JSON
print("\n")
```

# 5. Trabalhando com cabeçalhos (headers)

```python
print("5. Requisição com Cabeçalhos Personalizados")
headers = {'User-Agent': 'MeuApp/1.0', 'Accept': 'application/json'}
response = requests.get('https://httpbin.org/get', headers=headers)

print("Cabeçalhos enviados:")
# Convertendo para dicionário para melhor visualização
print(dict(response.request.headers))
print("\nResposta:")
print(json.dumps(response.json(), indent=4))  # Melhor formatação para JSON
print("\n")

```

# 6. Acessando cabeçalhos da resposta


```python
print("6. Acessando Cabeçalhos da Resposta")
response = requests.get('https://httpbin.org/get')

print("Cabeçalhos da resposta:")
# Usando dict comprehension para melhor visualização
headers_dict = {k: v for k, v in response.headers.items()}
print(json.dumps(headers_dict, indent=4))
print("\n")
```

# 7. Trabalhando com cookies


```python
print("7. Trabalhando com Cookies")
# Corrigindo para usar a API correta do httpbin
response = requests.get('https://httpbin.org/cookies/set/nome/valor', allow_redirects=True)

print("Cookies recebidos:")
print(response.cookies.get_dict())
print("Resposta:")
print(json.dumps(response.json(), indent=4))  # Adicionando visualização da resposta
print("\n")
```

# 8. Enviando cookies em uma requisição

```python

print("8. Enviando Cookies")
cookies = {'session_id': '12345', 'usuario': 'joao123'}
response = requests.get('https://httpbin.org/cookies', cookies=cookies)

print("Resposta:")
print(json.dumps(response.json(), indent=4))  # Melhor formatação para JSON
print("\n")
```

# 9. Lidando com redirecionamentos

```python

print("11. Tratando Exceções")
try:
    response = requests.get('https://site-que-nao-existe.com', timeout=2)  # Adicionando timeout
except requests.exceptions.ConnectionError:
    print("Erro de conexão: não foi possível conectar ao servidor.")
    print("Dica: Verifique se a URL está correta e se você tem conexão com a internet.")
except requests.exceptions.Timeout:
    print("Timeout: o servidor demorou muito para responder.")
    print("Dica: Tente aumentar o valor do timeout ou verificar sua conexão.")
except requests.exceptions.RequestException as e:
    print(f"Erro na requisição: {e}")
    print("Dica: RequestException é a classe base para todas as exceções do Requests.")
print("\n")
```

```python
import requests
import pandas as pd
import time
from tqdm import tqdm  # Para mostrar o progresso

# Obter todos os estados
estados_url = "https://servicodados.ibge.gov.br/api/v1/localidades/estados"
estados_response = requests.get(estados_url)
estados = estados_response.json()

```
