# Questionário sobre Apache Spark

Este notebook contém 10 questões objetivas sobre o Apache Spark baseadas no conteúdo do notebook "1.Introdução ao Apache Spark.ipynb".

## Questão 1: O que é o Apache Spark?

a) Um sistema de gerenciamento de banco de dados relacional

b) Um framework de processamento distribuído de código aberto projetado para computação rápida em grandes conjuntos de dados

c) Uma linguagem de programação específica para análise de dados

d) Um sistema operacional para servidores de big data

**Resposta correta: b**

## Questão 2: Qual é uma das principais vantagens do Spark em relação ao Hadoop MapReduce tradicional?

a) O Spark é mais antigo e estável

b) O Spark só funciona com dados estruturados

c) O Spark processa dados em memória, tornando-o significativamente mais rápido

d) O Spark só suporta a linguagem Java

**Resposta correta: c**

## Questão 3: Quais linguagens são suportadas pelas APIs do Spark?

a) Apenas Java e Scala

b) Apenas Python e R

c) Python, Java, Scala e R

d) C++, C# e JavaScript

**Resposta correta: c**

## Questão 4: O que significa a sigla RDD no contexto do Spark?

a) Rapid Data Distribution

b) Resilient Distributed Dataset

c) Relational Database Driver

d) Reactive Data Design

**Resposta correta: b**

## Questão 5: Qual componente é responsável por converter o código do usuário em tarefas que serão executadas pelos executores no Spark?

a) Cluster Manager

b) Executors

c) Driver Program

d) SparkContext

**Resposta correta: c**

## Questão 6: O que é o SparkContext?

a) Uma biblioteca para visualização de dados no Spark

b) Um componente que executa as tarefas atribuídas pelo driver

c) O ponto de entrada para qualquer funcionalidade Spark e representa a conexão com o cluster Spark

d) Um gerenciador de recursos no cluster

**Resposta correta: c**

## Questão 7: Quais são os processos iniciados em nós de trabalho que executam as tarefas atribuídas pelo driver no Spark?

a) Workers

b) Managers

c) Executors

d) Nodes

**Resposta correta: c**

## Questão 8: Qual é o ponto de entrada para programação Spark com a API Dataset e DataFrame?

a) SparkContext

b) SparkSession

c) SparkExecutor

d) SparkManager

**Resposta correta: b**

## Questão 9: Qual das seguintes NÃO é uma característica do PySpark?

a) Combina a facilidade do Python com o poder de processamento distribuído do Spark

b) Permite aos cientistas de dados usar bibliotecas Python populares em conjunto com Spark

c) Fornece APIs intuitivas para manipulação de dados complexos

d) É mais rápido que o Spark nativo em Scala para todas as operações

**Resposta correta: d**

## Questão 10: Qual dos seguintes gerenciadores de cluster pode ser usado com o Spark?

a) Apenas YARN

b) Apenas Kubernetes

c) YARN, Mesos, Kubernetes ou gerenciador standalone do Spark

d) Apenas Docker Swarm
