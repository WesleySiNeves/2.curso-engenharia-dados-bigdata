### 4. Operações básicas

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Operações aritméticas são elemento a elemento
print(a + b)  # [5 7 9]
print(a * b)  # [4 10 18]
print(a ** 2) # [1 4 9]

# Comparações também são elemento a elemento
print(a < 3)  # [True True False]
```

### 6. Estatísticas

```python
data = np.array([1, 2, 3, 4, 5])

print(np.mean(data))      # 3.0
print(np.median(data))    # 3.0
print(np.std(data))       # Desvio padrão
print(np.var(data))       # Variância
print(np.min(data))       # 1
print(np.max(data))       # 5
print(np.sum(data))       # 15
```

### 7. Reshape e redimensionamento

```python
# Reshape para 3x4
reshaped = arr.reshape(3, 4)
print(reshaped)
```

### 8. Broadcasting
O broadcasting permite que o NumPy trabalhe com arrays de diferentes formas quando realiza operações aritméticas:

```python
a = np.array([[1, 2, 3], [4, 5, 6]])  # forma (2, 3)
b = np.array([10, 20, 30])            # forma (3,)

# b é automaticamente expandido para forma (2, 3)
print(a + b)
# [[11 22 33]
#  [14 25 36]]
```

### Análise de dados de temperatura

#### Imagine que você trabalha para uma empresa agrícola e precisa analisar dados de temperatura para otimizar o plantio. Você tem medições diárias de temperatura ao longo de um ano.

```python
#explique o linspace
dias = 365
np.linspace(0, 2 * np.pi, dias)
```

```python
import numpy as np
import matplotlib.pyplot as plt  # Para visualização

# Simular dados diários de temperatura (em Celsius) para um ano
dias = 365
np.random.seed(42)  # Para reprodutibilidade
temperaturas = 20 + 10 * np.sin(np.linspace(0, 2*np.pi, dias)) + np.random.normal(0, 3, dias)

# Converter dados para array NumPy
temp_array = np.array(temperaturas)

# Análise básica
temperatura_media = np.mean(temp_array)
temperatura_maxima = np.max(temp_array)
temperatura_minima = np.min(temp_array)
temperatura_desv_padrao = np.std(temp_array)

print(f"Temperatura média: {temperatura_media:.1f}°C")
print(f"Temperatura máxima: {temperatura_maxima:.1f}°C")
print(f"Temperatura mínima: {temperatura_minima:.1f}°C")
print(f"Temperatura desvio padrão: {temperatura_desv_padrao:.1f}°C")
```

```python
# Identificar dias com temperatura acima de 30°C (potencialmente prejudiciais para certas culturas)
dias_quentes = np.where(temp_array > 30)[0]
percentual_dias_quentes = len(dias_quentes) / dias * 100
print(f"Percentual de dias com temperatura acima de 30°C: {percentual_dias_quentes:.1f}%")
```

```python
# Dividir em 12 meses usando array_split
temp_por_mes = np.array_split(temp_array, 12)
```

```python

# Calcular médias mensais
media_mensal = np.array([np.mean(mes) for mes in temp_por_mes])

# Visualizar os resultados
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
print("\nTemperaturas médias mensais:")
for mes, temp in zip(meses, media_mensal):
    print(f"{mes}: {temp:.1f}°C")
```

## Identificar o melhor mês para plantio (exemplo: temperatura ideal entre 18-25°C)

```python

ideal_planting_condition = np.abs(media_mensal - 21.5)  # Distância da temp. ideal (21.5°C)
melhor_mes_plantio = meses[np.argmin(ideal_planting_condition)]
print(f"\nMelhor mês para plantio: {melhor_mes_plantio}")

# Detectar períodos de seca (exemplo: 7 dias consecutivos acima de 28°C)
periodos_seca = []
contador = 0

for i, temp in enumerate(temp_array):
    if temp > 28:
        contador += 1
    else:
        if contador >= 7:
            periodos_seca.append((i-contador, i-1))
        contador = 0

# Verificar se o último período é de seca
if contador >= 7:
    periodos_seca.append((dias-contador, dias-1))

print(f"\nPeríodos de seca (7+ dias acima de 28°C): {len(periodos_seca)}")
for inicio, fim in periodos_seca[:3]:  # Mostrar apenas os 3 primeiros períodos
    print(f"Dias {inicio+1} a {fim+1}: {fim-inicio+1} dias")
```
