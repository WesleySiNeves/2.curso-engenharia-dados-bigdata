## 6. Tratamento de Dados Ausentes

```python
# Criar dados com valores ausentes
dados_faltantes = {
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [9, 10, 11, 12]
}
df_na = pd.DataFrame(dados_faltantes)
df_na
```

### Verificar dados ausentes

```python
# Verificar dados ausentes
print(df_na.isna())  # Retorna True onde há NaN
```

```python
print(df_na.isna().sum())  # Conta NaNs por coluna
```

```python
# Preencher dados ausentes
df_preenchido = df_na.fillna(0)  # Preenche com zeros

df_preenchido
```

```python
df_forward = df_na.ffill()  # Propaga o último valor válido
df_forward
```

```python
df_backward = df_na.bfill()  # Usa o próximo valor válido
df_backward
```

```python

# Remover linhas/colunas com dados ausentes
df_limpo = df_na.dropna()  # Remove linhas com qualquer NaN
df_limpo
```

```python
df_limpo_cols = df_na.dropna(axis=1)  # Remove colunas com qualquer NaN
df_limpo_cols
```

### 7. Agrupamento e Agregação

```python
# Dados de exemplo
dados_vendas = {
    'Data': pd.date_range('2023-01-01', periods=10),
    'Produto': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A', 'C'],
    'Vendedor': ['Maria', 'João', 'Maria', 'Pedro', 'João', 
                 'Maria', 'Pedro', 'João', 'Pedro', 'Maria'],
    'Quantidade': [10, 5, 8, 12, 7, 9, 11, 6, 15, 13],
    'Valor': [100, 150, 100, 200, 150, 100, 200, 150, 100, 200]
}
vendas = pd.DataFrame(dados_vendas)

vendas
```

```python
# Agrupamento simples
por_produto = vendas.groupby('Produto')['Quantidade'].sum()
por_produto
```

```python

# Agrupamento com múltiplas funções de agregação
estatisticas = vendas.groupby('Produto').agg({
    'Quantidade': ['sum', 'mean', 'count'],
    'Valor': ['sum', 'mean']
})

estatisticas
```

```python
# Agrupamento por múltiplas colunas
vendas_detalhadas = vendas.groupby(['Produto', 'Vendedor']).agg({
    'Quantidade': 'sum',
    'Valor': 'sum'
})


vendas_detalhadas
```

## 8. Mesclagem e Concatenação de DataFrames

```python
# Dados para junção
clientes = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo'],
})

clientes
```

```python
pedidos = pd.DataFrame({
    'ID_Cliente': [3, 1, 5, 2, 1, 3],
    'Produto': ['Mouse', 'Teclado', 'Monitor', 'CPU', 'HD', 'Webcam'],
    'Valor': [50, 100, 500, 1200, 300, 80]
})

pedidos
```

```python

# Junção (merge)
clientes_pedidos = pd.merge(clientes, pedidos, 
                            left_on='ID', right_on='ID_Cliente')
clientes_pedidos
```

### Diferentes tipos de junção

```python

# Diferentes tipos de junção
inner_join = pd.merge(clientes, pedidos, left_on='ID', right_on='ID_Cliente')
inner_join
```

```python
left_join = pd.merge(clientes, pedidos, left_on='ID', right_on='ID_Cliente', how='left')
left_join
```

```python
right_join = pd.merge(clientes, pedidos, left_on='ID', right_on='ID_Cliente', how='right')
right_join
```

```python
# Concatenação
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
concatenado_vert = pd.concat([df1, df2])  # Empilha verticalmente
concatenado_vert
```

```python
concatenado_horiz = pd.concat([df1, df2], axis=1)  # Concatena horizontalmente
concatenado_horiz
```

## 9. Pivô e Reestruturação

```python
# Criar dados de exemplo
dados_vendas_mensais = {
    'Ano': [2022, 2022, 2022, 2023, 2023, 2023] * 2,
    'Mês': [1, 2, 3, 1, 2, 3] * 2,
    'Produto': ['A'] * 6 + ['B'] * 6,
    'Vendas': [100, 110, 120, 150, 160, 170, 200, 210, 220, 250, 260, 270]
}
vendas_mensais = pd.DataFrame(dados_vendas_mensais)
```

```python
vendas_mensais
```

```python
# Tabela pivô
tabela_pivot = pd.pivot_table(
    vendas_mensais, 
    values='Vendas', 
    index=['Ano', 'Mês'], 
    columns='Produto', 
    aggfunc='sum'
)
tabela_pivot
```

```python

# De formato longo 
vendas = vendas_mensais.pivot(
    index=['Ano', 'Mês'],
    columns='Produto',
    values='Vendas'
)
vendas
```

## 10. Exemplo Prático: Análise de Vendas
Vamos aplicar esses conceitos em um exemplo prático:

```python
# Criar dados de vendas
np.random.seed(42)
datas = pd.date_range('2023-01-01', periods=100)
produtos = ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'HD Externo']
regioes = ['Norte', 'Sul', 'Leste', 'Oeste', 'Centro']

dados = {
    'Data': np.random.choice(datas, 1000),
    'Produto': np.random.choice(produtos, 1000),
    'Região': np.random.choice(regioes, 1000),
    'Quantidade': np.random.randint(1, 10, 1000),
    'Preço': np.random.choice([100, 50, 30, 200, 80], 1000)
}

df_vendas = pd.DataFrame(dados)
df_vendas
```

```python
df_vendas.head()
```

```python
df_vendas['Valor Total'] = df_vendas['Quantidade'] * df_vendas['Preço']
df_vendas['Mês'] = df_vendas['Data'].dt.month

df_vendas
```

```python

# Análise de vendas por produto
vendas_por_produto = df_vendas.groupby('Produto').agg({
    'Quantidade': 'sum',
    'Valor Total': 'sum'
}).sort_values('Valor Total', ascending=False)
vendas_por_produto
```

```python

# Análise de vendas por região e mês
vendas_regiao_mes = df_vendas.groupby(['Região', 'Mês']).agg({
    'Valor Total': 'sum'
}).unstack()
```

```python

# Identificar os dias com maiores vendas
vendas_por_dia = df_vendas.groupby('Data')['Valor Total'].sum().sort_values(ascending=False)
top_5_dias = vendas_por_dia.head(5)
top_5_dias
```

```python
# Calcular média móvel para visualizar tendência
vendas_diarias = df_vendas.groupby('Data')['Valor Total'].sum()
media_movel = vendas_diarias.rolling(window=7).mean()

# Criar um DataFrame a partir da série de vendas diárias
df_vendas_diarias = vendas_diarias.reset_index()
df_vendas_diarias.columns = ['Data', 'Valor Total']

# Adicionar a média móvel como uma nova coluna
df_vendas_diarias['mediamovel'] = media_movel.values

# Visualizar o resultado
df_vendas_diarias
```

```python

# Plotar os dados
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df_vendas_diarias['Data'], df_vendas_diarias['Valor Total'], label='Vendas Diárias')
plt.plot(df_vendas_diarias['Data'], df_vendas_diarias['mediamovel'], label='Média Móvel (7 dias)', color='red')
plt.title('Vendas Diárias e Média Móvel')
plt.xlabel('Data')
plt.ylabel('Valor Total')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
