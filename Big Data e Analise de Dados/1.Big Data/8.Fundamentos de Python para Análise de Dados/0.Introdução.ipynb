### Conceitos Fundamentais para Análise de Dados
#### Vetorização vs. Loops
A vetorização é um conceito essencial para trabalhar eficientemente com grandes volumes de dados:

```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
# Criando um array grande para o teste
data = np.random.rand(1000000)
```

```python
# Abordagem com loop (lenta)
start_time = time.time()
result_loop = []
for x in data:
    result_loop.append(x * 2)
loop_time = time.time() - start_time
print(f"Tempo com loop: {loop_time:.4f} segundos")
```

```python
# Abordagem vetorizada (rápida)
start_time = time.time()
result_vector = data * 2
vector_time = time.time() - start_time
print(f"Tempo vetorizado: {vector_time:.4f} segundos")
```

```python
# Calculando a diferença de velocidade
speedup = loop_time / vector_time
print(f"A abordagem vetorizada é {speedup:.1f}x mais rápida")
```

```python
# Visualizando a diferença
plt.figure(figsize=(10, 6))
plt.bar(['Loop', 'Vetorizado'], [loop_time, vector_time], color=['#ff9999', '#66b3ff'])
plt.title('Comparação de Performance: Loop vs. Vetorizado', fontsize=15)
plt.ylabel('Tempo (segundos)', fontsize=12)
plt.yscale('log')  # Escala logarítmica para melhor visualização
plt.grid(axis='y', alpha=0.3)

# Adicionando os valores de tempo nas barras
for i, v in enumerate([loop_time, vector_time]):
    plt.text(i, v*1.1, f"{v:.4f}s", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

### A vetorização permite:

#### Código mais limpo e conciso
#### Execução significativamente mais rápida
#### Melhor utilização de recursos computacionais
