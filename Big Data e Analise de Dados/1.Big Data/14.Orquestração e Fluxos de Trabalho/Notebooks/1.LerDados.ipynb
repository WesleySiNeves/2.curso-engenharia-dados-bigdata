## Importar dados da API

```python
import requests
url = 'https://dadosabertos.camara.leg.br/api/v2/deputados'
itens = []
resp = requests.get(url).json()
for d in resp['dados']:
    if d['siglaUf'] == 'DF':
        itens.append(d['id'])
        print(f"https://dadosabertos.camara.leg.br/api/v2/deputados/{d['id']}")
print(itens)
            
```
